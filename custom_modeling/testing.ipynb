```python
import matplotlib.pyplot as plt


output_dir = '/Users/jkember/Documents/nia_health/NHANES/biological_age/data/output_male/egfr'

delta_age_df = pd.read_csv(f'{output_dir}/model/delta_age-array.csv', index_col=0)

age_values = delta_age_df.columns.astype(float)
feature_values = delta_age_df.index.astype(float)
delta_age_grid = delta_age_df.to_numpy()

# Plot.
fig, ax = plt.subplots(figsize=(10,6))

c_lim = np.max(np.abs(delta_age_grid))
contour = ax.contourf(age_values, feature_values, delta_age_grid, levels=100, cmap='coolwarm', vmin=-c_lim, vmax=c_lim)
ax.contour(age_values, feature_values, delta_age_grid, levels=10, colors='w', linewidths=0.2)
ax.figure.colorbar(contour, ax=ax, label='$\Delta$ age', shrink=0.8)

# ax.plot(age_vals, interp_func(age_vals), linestyle='--', lw=3, color='k')

ax.set_xlabel('age')
```
